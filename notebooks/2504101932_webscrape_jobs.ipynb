```python
from jobspy import scrape_jobs
import pandas as pd
```

```python
jobs = scrape_jobs(
    site_name=["indeed", #"linkedin", 
               "glassdoor", 
               "google"],
    search_term="food science",
    google_search_term="food science jobs Singapore",
    location="Singapore",
    hours_old=17520,
    results_wanted=2000,
    country_indeed='Singapore',
    
    # linkedin_fetch_description=True # gets more info such as description, direct job url (slower)
    # proxies=["208.195.175.46:65095", "208.195.175.45:65095", "localhost"],
)
print(f"Found {len(jobs)} jobs")
```

```python
# another term
jobs_tech = scrape_jobs(
    site_name=["indeed", #"linkedin", 
               "glassdoor", 
               "google"],
    search_term="food technology",
    google_search_term="food tech jobs Singapore",
    location="Singapore",
    hours_old=17520,
    results_wanted=2000,
    country_indeed='Singapore'
)
print(f"Found {len(jobs_tech)} jobs")
```
